```python
import cv2
from PIL import Image
```

```python
img = cv2.imread('Lena.jpg')
Image.fromarray(img[..., ::-1])
```

```python
# utworzenie detektora SIFT o standardowych parametrach
sift = cv2.SIFT_create()
```

```python
# detekcja punktów kluczowych
kps = sift.detect(img, None)
print(len(kps), type(kps[0]), kps[0].pt)
```

```python
# wyświetlenie punktów kluczowych
img_kp = cv2.drawKeypoints(img, kps, None, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
Image.fromarray(img_kp[..., ::-1])
```
